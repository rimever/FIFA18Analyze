```python
#サンプルをランダムに分けるcross_validation
#import sklearn.cross_validation 
import sklearn.model_selection

#左辺にモデル構築用のXY,テスト用のX,Yを用意します。
#先ほど作成した、説明変数を第一引数に、目的変数を第二引数に置きます。
#train_test_splitを足すことを忘れないでください！！
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X,Y)
```

```python
X_test.shape

X_train.shape

X.shape
```

```python
#線形回帰モデルを作成
multi_lreg = LinearRegression()

#X,Y双方がtrain用のデータである点にご注意ください。
multi_lreg.fit(X_train, Y_train)
```

```python
#トレーニングに使用したデータセットでの精度
print(multi_lreg.score(X_train,Y_train))

#テストに使用したデータセットでの精度
print(multi_lreg.score(X_test,Y_test))

```

```python
from sklearn import metrics
from sklearn import linear_model

linreg = linear_model.LinearRegression()
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,Y)
linreg.fit(X_train, y_train)
```

```python
metrics.mean_squared_error(Y_train, linreg.predict(X_train))
```

```python
linreg.score(X_train,y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
metrics.mean_squared_error(Y_test, y_pred)
```

```python
metrics.mean_squared_error(Y_train, linreg.predict(X_train))
```

```python
# ラッソ回帰
lassoreg = linear_model.Lasso()
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,Y)
lassoreg.fit(X_train, y_train)
```

```python
lassoreg.score(X_train,y_train)
```

```python
y_pred = lassoreg.predict(X_test)
metrics.mean_squared_error(Y_test, y_pred)
```

```python
mse = metrics.mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

```python
# リッジ回帰アルゴリズム
ridgereg = linear_model.Ridge()
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,Y)
ridgereg.fit(X_train, y_train)
```

```python
metrics.mean_squared_error(Y_train, ridgereg.predict(X_train))
```

```python
ridgereg.score(X_train,y_train)
```

```python
y_pred = ridgereg.predict(X_test)
metrics.mean_squared_error(Y_test, y_pred)
```

```python
mse = metrics.mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

```python

#説明変数を縦(1)の列と指定して削除します！
X = result.drop("Goal", 1)

#Yに目的変数を入れます！
Y = result.Goal

ridgereg = linear_model.Ridge()
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,Y)
ridgereg.fit(X_train, y_train)
                  
print("metrics(train)=" + str(metrics.mean_squared_error(Y_train, ridgereg.predict(X_train))))
print("score=" + str(ridgereg.score(X_train,y_train)))
y_pred = ridgereg.predict(X_test)
print("metrics(test)=" + str(metrics.mean_squared_error(Y_test, y_pred)))
```

```python
import lightgbm as lgb
train_data_set = lgb.Dataset(X_train, y_train)
test_data_set = lgb.Dataset(X_test, y_test, reference=train_data_set)

params = {                                                                                               
    'boosting_type': 'gbdt',                                                                             
    'objective': 'regression_l2',                                                                           
    'metric': 'l2',                                                                             
    'num_leaves': 40,                                                                                    
    'learning_rate': 0.05,                                                                               
    'feature_fraction': 0.9,                                                                             
    'bagging_fraction': 0.8,                                                                             
    'bagging_freq': 5,   
    'lambda_l2': 2,
}                                                                                                        
                                                                                                         
gbm = lgb.train(params,                                                                                  
                train_data_set,                                                                               
                num_boost_round=200,                                                                      
                valid_sets=test_data_set,                                                                     
                early_stopping_rounds=10
                )   
```

```python
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

```python
metrics.mean_squared_error(Y_test, y_pred)
```

```python
y_pred
```

```python
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```
