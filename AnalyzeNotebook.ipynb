```python
# ガウス関数
def gauss(x, mu, s):
    return np.exp(-(x-mu)**2 / (2*s**2))

def gauss_func(w,x,x_min,x_max):
    m = len(w) - 1
    mu = np.linspace(x_min,x_max,m)
    s = mu[1] - mu[0]
    y = np.zeros_like(x)
    for j in range(m):
        y = y + w[j] * gauss(x,mu[j],s)
        
    y = y + w[m]
    return y

def mse_gauss_func(x,t,w,x_min,x_max):
    y = gauss_func(w,x,x_min,x_max)
    mse = np.mean((y-t)**2)
    return mse

def fit_gauss_func(x,t,m,x_min,x_max):
    mu = np.linspace(x_min,x_max,m)
    s = mu[1] - mu[0]
    n = x.shape[0]
    psi = np.ones((n, m+1))
    for j in range(m):
        psi[:,j] = gauss(x,mu[j],s)
    psi_T = np.transpose(psi)
    
    b = np.linalg.inv(psi_T.dot(psi))
    c = b.dot(psi_T)
    w = c.dot(t)
    return w

def show_gauss_func(w,x_min,x_max):
    xb = np.linspace(x_min, x_max, 100)
    y = gauss_func(w, xb,x_min,x_max)
    plt.plot(xb, y, c=[.5,.5,.5],lw=4)
    
```

```python
overall_data = data['Overall'] 
value_data = data['Value']
X = overall_data 
T = value_data 
x_min = np.min(overall_data) 
x_max = np.max(overall_data)
plt.figure(figsize=(4,4)) 
M=15
W = fit_gauss_func(X,T,M,x_min,x_max) 
plt.plot(X,T,marker='o',linestyle='None',color='cornflowerblue',markeredgecolor='black') 
show_gauss_func(W,x_min,x_max) 
plt.xlim(x_min,x_max) 
plt.grid(True) 
mse = mse_gauss_func(X,T,W,x_min,x_max) 
print('W=' + str(np.round(W,1))) 
print("SD ={0:.2f}".format(np.sqrt(mse))) 
plt.show()
```
