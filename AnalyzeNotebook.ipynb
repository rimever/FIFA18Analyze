```python
# ガウス関数
def gauss(x, mu, s):
    return np.exp(-(x-mu)**2 / (2*s**2))

def gauss_func(w,x,x_min,x_max):
    m = len(w) - 1
    mu = np.linspace(x_min,x_max,m)
    s = mu[1] - mu[0]
    y = np.zeros_like(x)
    for j in range(m):
        y = y + w[j] * gauss(x,mu[j],s)
        
    y = y + w[m]
    return y

def mse_gauss_func(x,t,w,x_min,x_max):
    y = gauss_func(w,x,x_min,x_max)
    mse = np.mean((y-t)**2)
    return mse

def fit_gauss_func(x,t,m,x_min,x_max):
    mu = np.linspace(x_min,x_max,m)
    s = mu[1] - mu[0]
    n = x.shape[0]
    psi = np.ones((n, m+1))
    for j in range(m):
        psi[:,j] = gauss(x,mu[j],s)
    psi_T = np.transpose(psi)
    
    b = np.linalg.inv(psi_T.dot(psi))
    c = b.dot(psi_T)
    w = c.dot(t)
    return w

def show_gauss_func(w,x_min,x_max):
    xb = np.linspace(x_min, x_max, 100)
    y = gauss_func(w, xb,x_min,x_max)
    plt.plot(xb, y, c=[.5,.5,.5],lw=4)
    
```

```python
# K分割交差検証
def kfold_gauss_func(x, t, m, k,x_min,x_max) :
    n = x.shape[0]
    mse_train = np.zeros(k)
    mse_test = np.zeros(k)
    for i in range(0,k):
        x_train = x[np.fmod(range(n), k) != i]
        t_train = t[np.fmod(range(n), k) != i]
        x_test = x[np.fmod(range(n), k) == i]
        t_test = t[np.fmod(range(n), k) == i]

        wm = fit_gauss_func(x_train, t_train, m,x_min,x_max) 
        
        mse_train[i] = mse_gauss_func(x_train, t_train, wm,x_min,x_max) 
        mse_test[i] = mse_gauss_func(x_test, t_test, wm,x_min,x_max) 
    return mse_train,mse_test
```

```python
M = range(2,50)

K = 16

Cv_Gauss_train = np.zeros((K, len(M)))
Cv_Gauss_test = np.zeros((K, len(M)))
for i in range(0, len(M)):
    Cv_Gauss_train[:,i],Cv_Gauss_test[:,i] = kfold_gauss_func(X,T,M[i],K,x_min,x_max)
    
mean_Gauss_train = np.sqrt(np.mean(Cv_Gauss_train, axis=0))
mean_Gauss_test = np.sqrt(np.mean(Cv_Gauss_test, axis=0))

plt.figure(figsize=(4,3))
plt.plot(M, mean_Gauss_train, marker='o', linestyle='-',
         color='k', markerfacecolor='w', label='training')
plt.plot(M, mean_Gauss_test, marker='o', linestyle='-',
         color='cornflowerblue',markeredgecolor='black',label='test')
plt.legend(loc='upper left', fontsize=10)
plt.grid(True)
plt.show()                  
```

```python
def model_A(x,w):
    y = w[0] - w[1] * np.exp(-w[2] * x)
    return y

def show_model_A(w):
    xb = np.linspace(x_min,x_max,100)
    y = model_A(xb,w)
    plt.plot(xb, y, c =[.5,.5,.5], lw=4)
    
def mse_model_A(w,x,t):
    y = model_A(x,w)
    mse = np.mean((y-t)**2)
    return mse

from scipy.optimize import minimize

# パウェルアルゴリズム
def fit_model_A(w_init, x, t):
    res1 = minimize(mse_model_A, w_init,args=(x,t), method="powell")
    return res1.x

plt.figure(figsize=(4,4))
W_init = [100,0,0]
W = fit_model_A(W_init,X,T)
print("w0 = {0:.1f},w1={1:.1f},w2={2:.1f}".format(W[0],W[1],W[2]))

show_model_A(W)
plt.plot(X,T,marker='o',linestyle='None',
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(x_min,x_max)
plt.grid(True)
mse = mse_model_A(W,X,T)
print("SD={0:.2f} cm".format((np.sqrt(mse))))
plt.show()
```
